```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('messy_orders_550.csv')
df
```

```python
df.info()
```

```python
df.duplicated('order_id')
df = df.drop_duplicates('order_id')
df
```

```python
df['customer_name'] = df['customer_name'].fillna('Unknown')
df['customer_name'] = df['customer_name'].str.strip().str.title()
df
```

```python
df['email'] = df['email'].str.lower().str.strip()
df['email'] = df['email'].fillna('Not Given')
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
df['order_date'] = df['order_date'].fillna(pd.Timestamp('2024-01-01'))
```

```python
df['delivery_date'] = pd.to_datetime(df['delivery_date'], errors='coerce')
df['delivery_date'] = df['delivery_date'].fillna(pd.Timestamp('2024-01-10'))
```

```python
df['product'] = df['product'].str.strip().str.title()
```

```python
for i in ['quantity', 'price', 'total']:
    i_mean = df[i].mean()
    df[i] = df[i].fillna(i_mean).astype('int')
```

```python
df['payment_mode'] = df['payment_mode'].str.strip().str.title()
```

```python
df['city'] = df['city'].str.strip().str.title()
```

```python
df['feedback'] = df['feedback'].str.strip().str.title()
i_mode = df['feedback'].mode()[0]
df['feedback'] = df['feedback'].fillna(i_mode)
```
