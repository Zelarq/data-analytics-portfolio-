```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('project2.csv')
df
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df['Annual_Income']=df['Annual_Income'].fillna(df['Annual_Income'].median())
df['Satisfaction_Rating']=df['Satisfaction_Rating'].fillna(df['Satisfaction_Rating'].median())
df['Credit_Score']=df['Credit_Score'].fillna(df['Credit_Score'].median())
```

```python
df.info()
```

```python
df.head(10)
```

```python
df['Age']=df['Age'].round()
```

```python
df['Gender'].unique()
```

```python
df['Is_Prime_Member']=df['Is_Prime_Member'].replace({0,1},{'No','Yes'})
```

```python
df["Age_Group"] = pd.cut(df["Age"], bins=[0,18,35,50,80], labels=["Teen","Young","Adult","Senior"])
```

```python
df["Gender"].value_counts().plot(kind="pie", autopct='%1.1f%%')
plt.show()
```

Conclusion

* The dataset shows that customer spending does not strongly depend on age, gender, income, or Prime membership.
