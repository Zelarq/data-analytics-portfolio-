```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df =  pd.read_csv("ecommerce_orders.csv")
df.head()
```

```python
df.info()
```

```python
df  = df.drop_duplicates()
```

```python
#Which cities have the highest number of orders?
plt.figure(figsize=(7,3))
sns.countplot(df,x='city')
plt.show()
```

Findings : Kochi has the most customer orders while Chennai has the least, with the other cities showing moderate and similar demand.


```python
#How many orders were Delivered, Returned, Shipped, and Cancelled?
plt.figure(figsize=(7,3))
sns.countplot(df,x='status')
plt.show()
```

Findings : Returned orders are the highest, while delivered, shipped, and cancelled orders all occur at slightly lower but similar levels.


```python
#Which category (Electronics, Mobiles, Accessories) generated the most revenue?
plt.figure(figsize=(7,3))
Revenue = df['price'] * df['quantity']
sns.barplot(df,x='category',y=Revenue)
plt.show()
```

Findings : Mobiles generated the highest revenue, with Electronics and Accessories earning slightly less but still at similar levels.


```python
plt.figure(figsize=(7,3))
sns.countplot(df,x='product')
plt.show()
```

Findings : Laptops had the highest number of orders, while Monitor received the least, with all other products falling in a fairly similar mid-range.


```python
#What is the distribution of product prices?
plt.figure(figsize=(7,3))
sns.histplot(df,x='price')
plt.show()
```

Findings : Product prices are spread out evenly across the entire range, showing no strong concentration around any specific price level.


```python
p_cnt = df['status'].value_counts()
p_cnt.index
```

```python
#What percentage of orders are Delivered vs Returned vs Cancelled?
plt.pie(x=p_cnt.values,labels=p_cnt.index,autopct='%0.f%%')
plt.show()
```

Findings : Returned orders make up the largest share, while delivered, shipped, and cancelled orders each take up nearly equal portions of the total.


# Conclusion

Kochi and laptops lead in orders, mobiles generate the most revenue, order statuses are fairly balanced with slightly more returns, and price has no influence on how much customers buy.
