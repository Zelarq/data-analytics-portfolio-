```python
#plot countplot
plt.figure(figsize=(12,6))
sns.countplot(df,x='Category')
plt.xticks(rotation=60)
plt.show()
```

The countplot shows that Games is the most popular category with around 1,950 apps, followed by Family with about 1,100. Most other categories have fewer apps, and Medical has the least. This means developers focus more on games and entertainment since they attract more users and profits.


```python

#pie chart showing distribution of top 5 categories
plt.figure(figsize=(4,4))
df['Category'].value_counts().head(5).plot.pie(autopct='%1.1f%%')
plt.show()
```

Family apps lead the Play Store with 40.4%, followed by Games at 23.4% and Tools at 17.3%. This shows that entertainment and lifestyle apps dominate, while categories like Business and Medical are less common. Developers have the most chances — and competition — in Family and Games apps.


```python
plt.figure(figsize=(4,5))
sns.heatmap(df.corr(numeric_only=True))
```

```python
plt.figure(figsize=(4,5))
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.show()
```

The heatmap shows almost no correlation (0.065) between app ratings and the number of reviews. This means popular apps don’t always have better ratings — app quality is independent of how many people review it.


```python
#facet grid
g=sns.FacetGrid(df,col='Type')
g.map(plt.hist,'Rating')
plt.show()
```

The facet grid shows that most free apps have excellent ratings around 4.5 stars, while paid apps are fewer but have similar rating patterns. This means both free and paid apps maintain good quality, though the Play Store is mostly filled with free apps.


```python
#scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='Reviews',y='Rating',hue='Type')
plt.show()
```

The scatter plot shows that most free and paid apps have similar ratings with only a few reviews. Very few apps get a huge number of reviews, meaning many apps are well-rated but not very popular.


```python
#line plot
plt.figure(figsize=(10,6))
sns.lineplot(data=df,x='Reviews',y='Rating',hue='Type')
plt.show()
```

The line plot shows that app ratings remain steady around 4.0–4.5 stars for both free and paid apps, regardless of review count. Apps with fewer reviews vary more, but ratings stabilize as reviews increase — proving that app quality stays consistent no matter how popular it is.


```python
#histogram
plt.figure(figsize=(8,6))
sns.histplot(df['Rating'],bins=10)
plt.show()
```

Most apps on the Google Play Store have high ratings between 4.5 and 5.0 stars, showing strong user satisfaction. Very few apps have low ratings, indicating that poor-quality apps don’t last long on the platform.


```python

plt.figure(figsize=(4,4))
sns.boxplot(data=df,x='Type',y='Rating')
plt.show()
```

* When we compared free apps to paid apps, we found that free apps have a median rating of about 4.1 stars while paid apps have about 4.0 stars, which means they get almost the same ratings from users
* This shows us that choosing the right category is much more important than deciding whether to make an app free or paid!

```python
plt.figure(figsize=(15,4))
sns.boxplot(data=df,x='Category',y='Rating')
plt.xticks(rotation=90)
plt.show()
```

The category boxplot shows that most app categories have similar average ratings around 4.0–4.2 stars, meaning overall quality is consistent. However, categories like Music, Tools, and Games have more outliers with low ratings, while Education and Books have higher, steadier ratings. This means some categories vary more in quality than others.


CONCLUSION:

The overall analysis shows that success in the Play Store depends more on choosing the right category and maintaining good quality than on whether an app is free or paid. Games and Family apps dominate the store, but they also have the most competition. Most apps, both free and paid, have ratings above 4 stars, proving that users value quality. Developers should focus on making reliable and useful apps because good quality apps get high ratings even without millions of downloads. For users, this means that most apps available on the Play Store are well-rated and offer a good experience.
