```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("supermarket.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Gender'].mode()
```

```python
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df['Rating']=df['Rating'].fillna(df['Rating'].mean())
```

```python
df['Product Line']=df['Product Line'].fillna("Unknown")
```

```python
df.info()
```

```python
df['Payment'].unique()
```

```python
df['Customer Type']=df['Customer Type'].fillna(df['Customer Type'].mode()[0])
df['Payment']=df['Payment'].fillna(df['Payment'].mode()[0])
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['month']=df['Date'].dt.month_name()
```

```python
df
```

```python
plt.figure(figsize=(15,3))
sns.boxplot(x=df['Total'])
plt.xticks(range(0,1100,20),rotation=90)
plt.show()
```

```python
v=df['Payment'].value_counts()
plt.pie(v.values,labels=v.index,autopct='%1.1f%%')
plt.show()
```

```python
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()
```
